```python
# Input: All trials as one string, with linebreaks as in an output
# Output: A list of trials, with linebreaks removed
def split_into_trials(all_trials):
    lines = all_trials.split('\n')
    
    trials = []
    curr_trial = ''
    open_parens_minus_close_parens = 0
    for line in lines:
        curr_trial += line
        open_parens = line.count('(')
        close_parens = line.count(')')
        open_parens_minus_close_parens += open_parens - close_parens
        if open_parens_minus_close_parens == 0:
            trials.append(curr_trial)
            curr_trial = ''
            
    # Formatting
    for i in range(len(trials)):
        trials[i] = trials[i][:-2]
    
    return trials
```

```python
# Input: A trial
# Output: (path_length, num_expansions)    if a path was found
#         (None, num_expansions)           otherwise
def parse_trial(trial):
    num_expansions = int(trial.split(' ')[-1])
    try:
        path = trial[trial.index('[')+1:trial.index(']')]
        length = 0
        stack = []
        for i in path:
            if i == '(':
                stack.append(None)
            elif i == ')':
                stack.pop()
                if not stack:
                    length += 1

        return (length, num_expansions)
    except ValueError:        
        return (None, num_expansions)
```

```python
# Input: A list of trials
# Output: The same list, where each trial is replaced by its (length, num_expansions)
def get_data(trials):
    return [parse_trial(trial) for trial in trials]
```

```python
# Input: A list of pairs of the form (length, num_expansions)
# Output: ([sum_of_path_lengths,total_node_expansions],num_successes)
def parse_data_pairs(data_pairs, cap):
    good_data_pairs = [pair for pair in data_pairs if pair[0] != None and pair[1]<=cap]
    num_pairs = len(good_data_pairs)
    totals = [sum(x) for x in zip(*good_data_pairs)]
    return (totals, num_pairs)
```

```python
# Input: The output to be parsed
# Output: ([sum_of_path_lengths,total_node_expansions],num_successes)
def parse_output(output, cap):
    trials = split_into_trials(output)
    data_pairs = get_data(trials)
    return parse_data_pairs(data_pairs, cap)
```
